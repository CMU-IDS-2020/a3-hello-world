```python
import pandas as pd
```

```python
df = pd.read_csv("happy_map_filled.csv")
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.columns =['Country or Region', 'latitude', 'longitude', 'Rank', 'Score', 'GDP', 'Family',
       'Health', 'Freedom', 'Generosity', 'Corruption', 'Year']
```

```python
df.describe()
```

```python
count_df = df.groupby("Country or Region").agg({"Country or Region":"count"})
count_df[count_df["Country or Region"] != 5]
```

```python
df["Country or Region"].replace({"Hong Kong S.A.R., China":"Hong Kong",
                                  "Somaliland region": "Somaliland Region",
                                  "Taiwan Province of China":"Taiwan",
                                  "Trinidad & Tobago": "Trinidad and Tobago"}, inplace= True)
```

```python
count_df = df.groupby("Country or Region").agg({"Country or Region":"count"})
count_df[count_df["Country or Region"] != 5]
```

```python
df = df.drop(['latitude', 'longitude'], axis=1)
```

```python
for year in [2015,2016,2017,2018,2019]:
    df[df.Year == year].to_csv(str(year) + "_map.csv", index=False)
```
